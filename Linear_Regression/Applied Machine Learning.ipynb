# 1 Please plot the noisy data and the polynomial you found (in the same figure). You can use any value of m selected from 2, 3, 4, 5, 6.




```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
plt.show()
```

```python
# Finding Coefficients

co_efficients= np.polyfit(x.flatten(),y.flatten(),2) # Finding Coefficients
print('Coefficients',co_efficients) # Printing Co-Efficients
a = len(x) #count of x
print ("X co-ordinates {} count{}".format(x,a)) # Printing x Co-ordinates
b = len(y)  #count of y
print ("Y co-ordinates {} count{}".format(y,b)) # Printing y Co-ordinates
polynomial= np.poly1d(co_efficients)
plt.plot(x.flatten(),y.flatten(),'b.',x.flatten(),polynomial(x.flatten()),'or')

```

# 2 Plot MSE versus order m for m = 1, 2, 3, 4, 5, 6, 7, 8. Identify the best choice of order m
